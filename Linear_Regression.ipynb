linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
print("X values : \n",x)
```

```python
print("Y values : \n",y)
```

```python
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
```

```python
plt.scatter(x, y, color='blue', label = "Data Points")
plt.plot(x, y_pred, color='red', linewidth=2, label = "Regression Line")
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('LINEAR REGRESSION')
plt.show()
```

```python
print("Intercept: ", model.intercept_)
print("Slope: ", model.coef_)
```

```python
#multiple linear regression
q = 5 * np.random.rand(50, 1)
r = 3 * np.random.rand(50, 1)
s = 2 * np.random.rand(50, 1)
t = 4 + 2 * q + 3 * r + 1 * s + np.random.randn(50, 1)
model.fit(np.hstack((q, r, s)), t)
t_pred = model.predict(np.hstack((q, r, s)))
plt.scatter(t, t_pred, color='purple', label = "Data Points")
plt.plot(t, t_pred, color='pink', linewidth=2, label = "Regression Line")
plt.legend()
plt.xlabel('t')
plt.ylabel('t_pred')
plt.title('MULTIPLE LINEAR REGRESSION')
plt.show()
```

```python
print("Intercept: ", model.intercept_)
print("Slope: ", model.coef_)
```
